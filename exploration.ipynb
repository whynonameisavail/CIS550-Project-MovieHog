```python
credits_path = '/content/drive/MyDrive/CIS550_GroupProject/Movies_Metadata/credits.csv'
credits = pd.read_csv(credits_path)
credits.head()
```

```python
import ast 
credits['ast_cast'] = credits['cast'].apply(lambda x: ast.literal_eval(str(x)))
credits['ast_crew'] = credits['crew'].apply(lambda x: ast.literal_eval(str(x)))
```

```python
credits
```

```python
credits['ast_crew'][0]
```

```python
credits['ast_cast'][1][0]
```

```python
all_movies_cast = []
for i in range(len(credits)) :
  id = credits.loc[i, 'id']
  cast_list = credits.loc[i, 'ast_cast']
  m = len(cast_list)
  movie_cast = []
  for cast in cast_list: 
    name = cast['name']
    gender = cast['gender']
    character = cast['character']
    movie_cast.append([id, name, gender, character])
  all_movies_cast.extend(movie_cast)
all_movies_cast
```

```python
df_cast = pd.DataFrame(all_movies_cast, columns = ['movie_id', 'cast_name', 'gender', 'character'])
df_cast
```

```python
df_cast.to_csv (r'/content/drive/MyDrive/CIS550_GroupProject/Movies_Metadata/cast.csv', index = False, header=True)
```

```python
all_movies_crew = []
for i in range(len(credits)) :
  id = credits.loc[i, 'id']
  crew_list = credits.loc[i, 'ast_crew']
  m = len(crew_list)
  movie_crew = []
  for crew in crew_list: 
    name = crew['name']
    gender = crew['gender']
    job = crew['job']
    department = crew['department']
    movie_crew.append([id, name, gender, job, department])
  all_movies_crew.extend(movie_crew)
all_movies_crew
```

```python
df_crew = pd.DataFrame(all_movies_crew, columns = ['movie_id', 'cast_name', 'gender', 'job', 'department'])
print(df_crew)
df_crew.to_csv (r'/content/drive/MyDrive/CIS550_GroupProject/Movies_Metadata/crew.csv', index = False, header=True)
```

```python
keywords_path = '/content/drive/MyDrive/CIS550_GroupProject/Movies_Metadata/keywords.csv'
keywords = pd.read_csv(keywords_path)
keywords.head()
```

```python
keywords['ast_keywords'] = keywords['keywords'].apply(lambda x: ast.literal_eval(str(x)))
keywords.head()
```

```python
keywords['ast_keywords'][0]
```

```python
all_movies_keywords = []
for i in range(len(keywords)) :
  id = keywords.loc[i, 'id']
  keywords_list = keywords.loc[i, 'ast_keywords']
  m = len(keywords_list)
  movie_keywords = []
  for keyword in keywords_list: 
    key = keyword['name']
    movie_keywords.append([id, key])
  all_movies_keywords.extend(movie_keywords)
all_movies_keywords
```

```python
df_keywords = pd.DataFrame(all_movies_keywords, columns = ['movie_id', 'tag'])
print(df_keywords)
df_keywords.to_csv (r'/content/drive/MyDrive/CIS550_GroupProject/Movies_Metadata/tags.csv', index = False, header=True)
```

```python
meta_path = '/content/drive/MyDrive/CIS550_GroupProject/Movies_Metadata/ratings_small.csv'
ratings = pd.read_csv(meta_path)
ratings.head()
```

```python
meta_path = '/content/drive/MyDrive/CIS550_GroupProject/Movies_Metadata/movies_metadata.csv'
meta = pd.read_csv(meta_path)
meta.head()
```

```python
meta.columns    
```

```python
from datetime import datetime
dt_str = meta['release_date'][0]
print(dt_str)
date_time_obj = datetime.strptime(dt_str, '%Y-%m-%d')
```

```python
from datetime import datetime
all_movies_meta = []
for i in range(len(meta)) :
  id = meta.loc[i, 'id']
  dt_str = meta.loc[i, 'release_date']
  #print(dt_str)
  try:
    rel_date = datetime.strptime(dt_str, '%Y-%m-%d')
  except:
    rel_date = pd.NaT
  title = meta.loc[i, 'original_title']
  revs = meta.loc[i, 'revenue']
  try: 
    budget = float(meta.loc[i, 'budget'])
  except ValueError: 
    budget = pd.NA
  coll = meta.loc[i, 'belongs_to_collection']
  overview = meta.loc[i, 'overview']
  tagline = meta.loc[i, 'tagline']
  runtime = meta.loc[i, 'runtime']
  lang = meta.loc[i, 'original_language']
  imdb_id = meta.loc[i, 'imdb_id']
  vote = meta.loc[i, 'vote_average']
  all_movies_meta.append([id, title, revs, rel_date, budget, overview, tagline, runtime, lang, imdb_id, vote, coll])
#all_movies_meta
```

```python
all_movies_meta
```

```python
df_meta = pd.DataFrame(all_movies_meta, columns = ['movie_id', 'title', 'revs', 'release_date', 'budget', 'overview', 'tagline', 'runtime', 'lang', 'imdb_id', 'vote'])
print(df_meta)
df_meta.to_csv (r'/content/drive/MyDrive/CIS550_GroupProject/Movies_Metadata/meta_db.csv', index = False, header=True)
```

```python
all_movies_genres = []
for i in range(len(meta)) :
  id = meta.loc[i, 'id']
  genres = ast.literal_eval(str(meta.loc[i, 'genres']))
  for genre in genres:
    all_movies_genres.append([id, genre['name']])
all_movies_genres
```
