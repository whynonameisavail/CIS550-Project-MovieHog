```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt # plotting
import numpy as np # linear algebra
import os # accessing directory structure
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import json
```

```python
#parse all the utf16-encoded file
directory = os.fsencode('Downloads/Info/')
    
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    json_name = filename[:-3] + "json"
    if filename.endswith(".txt"): 
        #print(os.path.join(directory_in_str, filename))
        #currfile = os.path.join(directory_in_str, filename)
        file1 = open('Downloads/Info/' + filename, 'r', encoding = 'utf-16-le')
        file2 = open('Downloads/tempfile/' + filename, 'w', encoding = 'utf-8')
        file3 = open('Downloads/newparsed/' + json_name, 'w', encoding = 'utf-8')
        for line in file1.readlines():
            if not (line.startswith('  "_id"')):
                if (line.startswith("}")):
                    file2.write(line + ",\n")
                elif (line.endswith('\\",')):
                    line = line[:-4] + ',"'
                    file2.write(line)
                else:
                    file2.write(line)
        file2.close()
        file1.close()

        file2 = open('Downloads/tempfile/' + filename, 'r')
        data2 = file2.read()
        size = len(data2)
        data2 = '[' + data2[1:size - 2] + ']'
        file3.write(data2)
        file3.close()
        continue
    else:
        continue
            
```
